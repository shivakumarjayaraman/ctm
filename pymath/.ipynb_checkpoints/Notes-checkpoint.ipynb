```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fact(n) :
    if (n <= 1) : return 1
    return n * fact(n-1)

def sin_expansion(x, n):
    val = 0
    sign = -1
    for i in range(1, n+1, 2) :
        sign = sign * -1
        val += sign * (x**i)/fact(i)
    return val

def cos_expansion(x, n) : 
    val = 0
    sign = 1
    for i in range(0, n+1, 2) :
        val += sign * (x**i)/fact(i)
        sign *= -1
    return val

def exp_exp(x, n) :
    val = 0
    for i in range(n+1) :
        val += (x**i)/fact(i)
    return val
```

```python
## Plot sin curve
plt.plot(np.zeros(20), np.arange(-1, 1, 0.1))
plt.plot(np.arange(-25, 25), np.zeros(50))

points = np.array(list((x, sin_expansion(x, 100)) for x in np.arange(-25, 25, 0.01)))
plt.plot(points[:, 0], points[:, 1])

```

```python
## plot cosine curve
plt.plot(np.arange(-25, 25), np.zeros(50))
plt.plot(np.zeros(20), np.arange(-1, 1, 0.1))

points = np.array(list((x, cos_expansion(x, 100)) for x in np.arange(-25, 25, 0.01)))
plt.plot(points[:, 0], points[:, 1])
```

```python
## plot exponential function curve
points = np.array(list((x, exp_exp(x, 100)) for x in np.arange(0, 25, 0.01)))
plt.plot(points[:, 0], points[:, 1])
plt.plot(np.arange(30), np.zeros(30))
```

```python
## Preparing image 

import PIL.Image as Img

im = Img.open("SPP.jpg")
A = np.array(im)
m, n = A.shape
avr = A.sum()/(m*n)

L = np.zeros(256)
for i in range(m) :
    for j in range(n): 
        L[A[i, j]] += 1
        
plt.plot(range(256), L)
plt.plot([avr, avr], [0, L.max()])

print (A)

```

```python
from numpy import array

A = array(Img.open("SPP.jpg"))
B = A < 100
m, n = B.shape
h = max(m, n)
x0, y0 = m/h, n/h

def f(i, j):
    t1 = y0*(-1+2*j/(n-1))
    t2 = x0*(+1-2*i/(m-1))
    return (t1, t2)

P = [f(i, j) for i in range(m) for j in range(n) if B[i, j]]
with open('mypict.txt', 'w') as fd:
    fd.write(repr(P))
```

```python
with open('mypict.txt', 'r') as fd:
    P = eval(fd.read())

x, y = zip(*P)
plt.scatter(x, y, s=3)
plt.axis('scaled'), plt.xlim(-1, 1), plt.ylim(-1, 1), plt.show()
```
