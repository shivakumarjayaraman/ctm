```python
## Draw a unit circle .. 
n = 50
def ePowerIPiByN(x, n) :
    return math.e**(x * 2j * math.pi/n)

points = np.array([ePowerIPiByN(x, n) for x in range(n+1)])
cartesian = np.array([(x.real, x.imag) for x in points])

plt.plot(cartesian[:, 0], cartesian[:, 1])
```

```python
## How do you rotate an image by an angle 
S = {2 + 2j, 3 + 2j, 1.75 + 1j, 2 + 1j, 2.25 + 1j, 2.5 + 1j, 2.75 + 1j, 3 + 1j, 3.25 + 1j}
Spoints = np.array([(x.real, x.imag) for x in S])

plt.plot(Spoints[:,0], Spoints[:,1], 'r*')

## rotation matrix
angle = 180
def d2r(n) :
    return math.pi * n/180

rotMat = np.array([[np.cos(d2r(angle)), -1*np.sin(d2r(angle))], [np.sin(d2r(angle)), np.cos(d2r(angle))]])
rotated = (rotMat@Spoints.T).T
#rotated = rotated + np.array([5, 0])
plt.plot(rotated[:,0], rotated[:,1], 'b*')
```
